# Useful functions

```python
def plot_many(List_names,s=1,p=0):
    l_p=[]
    l_s=[]
    for col in List_names :
        l_s.append(pd.Series(list(df_SAL[col]), index=list(df_SAL['PERPAI'])))
        l_p.append(pd.Series(list(df_PAT[col]), index=list(df_PAT['PERPAI'])))

    plt.subplots(1, 1, figsize=(25, 10))
    
    if s==1 & p==1 : 
        for i in range(len(l_s)) : 
            plt.plot(l_s[i],label=List_names[i]+'_SAL')
            plt.plot(l_p[i],label=List_names[i]+'_PAT')
            
    elif s==1 :
         for i in range(len(l_s)) : 
            plt.plot(l_s[i],label=List_names[i])
            
        
    elif p==1 :
         for i in range(len(l_p)) : 
            plt.plot(l_p[i],label=List_names[i])

    plt.legend()
    plt.show()
            
```

```python
def plot_many_grid(ax, List_names, s=1, p=0):
    l_s = []
    l_p = []
    for col in List_names:
        l_s.append(pd.Series(list(df_SAL[col]), index=list(df_SAL['PERPAI'])))
        l_p.append(pd.Series(list(df_PAT[col]), index=list(df_PAT['PERPAI'])))

    if s == 1 and p == 1:
        for i in range(len(l_s)):
            ax.plot(l_s[i], label=List_names[i] + '_SAL')
            ax.plot(l_p[i], label=List_names[i] + '_PAT')
    elif s == 1:
        for i in range(len(l_s)):
            ax.plot(l_s[i], label=List_names[i])
    elif p == 1:
        for i in range(len(l_p)):
            ax.plot(l_p[i], label=List_names[i])

    ax.legend()
```

```python
l_cols = df_SAL.columns
```

```python
L_prime = []
for col in df_SAL.columns :
    if col[:5].lower() == 'prime' or col[:2].lower() == 'p.' or  col[:2].lower() == 'pr' or col[:8].lower() == 'assiduit' or col[:6].lower() == 'panier' :
        L_prime.append(col)

l_cols_1 = [col for col in l_cols if col not in L_prime]
```

```python
L_Indemnité = []
for col in l_cols_1 :
    if col[:3].lower() == 'ind' : 
        L_Indemnité.append(col)
        
l_cols_2 = [col for col in l_cols_1 if col not in L_Indemnité]
```

```python
L_Heures_supp = []
for col in l_cols_2 :
    if col[:3].lower() == 'hrs' or  col[:6].lower() == 'heures': 
        L_Heures_supp.append(col)
        
l_cols_3 = [col for col in l_cols_2 if col not in L_Heures_supp]
```

```python
L_Retenue = []
for col in l_cols_3 :
    if col[:3].lower() == 'ret' or  col[:7].lower() == 'retenue': 
        L_Retenue.append(col)
        
l_cols_4 = [col for col in l_cols_3 if col not in L_Retenue]
```

```python
L_Cotisation = []
for col in l_cols_4 :
    if col[:10].lower() == 'cotisation' or col[:8].lower() == 'mutuelle' or  col[:9].lower() == 'marocaine': 
        L_Cotisation.append(col)
        
l_cols_5 = [col for col in l_cols_4 if col not in L_Cotisation]
```

```python
L_Conge = []
for col in l_cols_5 :
    if col[:5].lower() == 'congé': 
        L_Conge.append(col)
        
l_cols_6 = [col for col in l_cols_5 if col not in L_Conge]
```

```python
L_pret_avance = []
for col in l_cols_6 :
    if col[:4].lower() == 'prêt' or col[:4].lower() == 'pret' or col[:6].lower() == 'avance' : 
        L_pret_avance.append(col)
        
l_cols_7 = [col for col in l_cols_6 if col not in L_pret_avance]
```

# Calculation of payroll (Masse salariale)

```python
df_SAL_cotis_sal = df_SAL[L_Cotisation]
df_SAL_prime = df_SAL[L_prime]
df_PAT_prime = df_PAT[L_prime]
```

```python
df_payroll_sums = pd.DataFrame()
df_payroll_sums['PERPAI'] = df_SAL['PERPAI']
```

```python
df_payroll_sums['Cotis. Sal'] = df_SAL_cotis_sal.sum(axis=1)
df_payroll_sums['Primes_sal'] = df_SAL_prime.sum(axis=1)
df_payroll_sums['Primes_pat'] = df_PAT_prime.sum(axis=1)
#df_payroll_sums['Net à payer'] = df_PAT['Net à payer']
df_payroll_sums['Net à payer_SAL'] = df_SAL['Net à payer']
```

```python
df_payroll_sums['Masse salariale'] = df_payroll_sums.iloc[:,1:].sum(axis=1)
```
